```python
import bs4
from urllib.request import urlopen as ureq
from bs4 import BeautifulSoup as soup

myurl = "https://www.flipkart.com/mens-footwear/nike~brand/pr?sid=osp,cil"
# Opening up connection and grabbing page
uClient = ureq(myurl)
page_html = uClient.read()
uClient.close()

# html parsing
page_soup = soup(page_html,'html.parser')

#grabs names page1
names = page_soup.findAll("div",{"class":"_2LFGJH"})


filename = "product.csv"
f = open(filename, "w", encoding = "utf-8-sig")

headers = "Shoe_Name, New_Price, Old_Price, Discount_Offer\n"

f.write(headers)

i=0
j=0
k=0


for name in names:
    shoe_name = name.a["title"]
    
    disc_prices = page_soup.findAll("div",{"class":"_1vC4OE"})
    new_price = disc_prices[i].text
    i = i+1
    
    act_prices = page_soup.findAll("div",{"class":"_3auQ3N"})
    old_price = act_prices[j].text
    j = j+1
    
    discount_container = page_soup.findAll("div",{"class":"VGWI6T"})
    discount = discount_container[k].span.text
    k = k+1
    
    print("Shoe Name:" + shoe_name)
    print("New Price:" + new_price)
    print("Old Price:" + old_price)
    print("Discount Offer:" + discount)
    print()
    f.write(shoe_name+ "," +new_price.replace(",","") + "," +old_price.replace(",","") + "," +discount + "\n")

f.close()
```
